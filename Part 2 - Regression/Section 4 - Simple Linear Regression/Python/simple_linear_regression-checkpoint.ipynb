# Simple Linear Regression

## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
data_file = "./Salary_Data.csv"
df = pd.read_csv(data_file)

x = df.iloc[:, :-1].values
y = df.iloc[:, -1:].values
print(x)
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
print(x_train)
print(x_test)
```

```python
print(y_train)
print(y_test)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

## Predicting the Test set results

```python
print(y_test)
```

```python
y_pred = regressor.predict(x_test)
```

```python
print(y_train)
```

```python
print(regressor.predict(x_train))
```

## Visualising the Training set results

```python
plt.scatter(x_train, y_train, color="red")
plt.plot(x_train, regressor.predict(x_train), color="blue")
plt.title("Salary vs Expirence (Training set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(x_test, y_test, color="red")
# plt.plot(x_test, y_pred, color="blue")
plt.plot(x_train, regressor.predict(x_train))
plt.title("Salary vs Experience (test set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

#### predicting the salary of an employer with 12 years of expirence

```python
print(f"${regressor.predict([[2]])[0][0] :.2f}")
```

##### Get the coffient and the y-intercept of an employee with 12 years of experience Salary = Y-intercept + Slope(Years)

```python
print(f"y-intercept: {regressor.intercept_}")
print(f"slope/coffient : {regressor.coef_}")
```
